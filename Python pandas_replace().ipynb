# Pandas replace()

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Saurabh\\OneDrive\\Documents\\Fortune2.csv')
df
```

```python
df.replace(to_replace = 'Health', value = 'Finance')
```

```python
df.replace('Health','Finance')
```

```python
df.replace(5, 15)
```

```python
df.replace([1,2,3,4,5], 0)
```

```python
df.replace([1,2,3,4,5], [11,12,13,14,15])
```

```python
df.replace({'Companies':'Health'}, 'none')
```

```python
df.replace({'Companies':'Health', 'Profit':5274553}, 'none')
```

```python
df
```

```python
df.replace('[A-Za-z]', 0)
```

```python
df.replace('[A-Za-z]', 0, regex = True)
```

```python
df.replace({'Companies':'[A-Za-z]'}, 0, regex = True)
```

```python
df
```

```python
df.replace('Health', method = 'ffill') #to_replace()
```

```python
df.replace('Health', method = 'bfill') #to_replace()
```

```python
df.replace('Health', method = 'bfill', limit = 1 )
```

```python
df.replace('Health', method = 'bfill', limit = 2 )
```

```python
df
```

```python
df.replace('Health', 'money', inplace = True) #to_values(), value() 
df
```

```python
df.replace('Health', 0, inplace = True) #to_values(), value() 
df
```
