# Pandas fillna()

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Saurabh\\OneDrive\\Documents\\Fortune3.csv')
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Name':'none', 'Profit':0, 'Growth':0})
```

```python
df.fillna(method = 'ffill')
```

```python
df.fillna(method = 'pad')
```

```python
df.fillna(method = 'bfill') # same as 'backfill'
```

```python
df.fillna(method = 'ffill', axis = 0)
```

```python
df.fillna(method = 'ffill', axis = 1)
```

```python
df.fillna(method = 'bfill', axis = 0)
```

```python
df.fillna(method = 'bfill', axis = 1)
```

```python
df.fillna(0, limit = 1) # 0-value parameter
```

```python
df.fillna(0, limit = 2)
```

```python
df.fillna(method = 'ffill', limit = 1)
```

```python
df.fillna(method = 'ffill', limit = 2)
```

```python
df.fillna(5, inplace = True) # 5-value parameter
df
```
